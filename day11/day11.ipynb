### Find the two monkeys that inspected the most number of items. Multiply those two numbers (of inspections) together.

```python
# Functions for each Monkey

# Monkey 0:
#   Starting items: 89, 74
#   Operation: new = old * 5
#   Test: divisible by 17
#     If true: throw to monkey 4
#     If false: throw to monkey 7

# Monkey 1:
#   Starting items: 75, 69, 87, 57, 84, 90, 66, 50
#   Operation: new = old + 3
#   Test: divisible by 7
#     If true: throw to monkey 3
#     If false: throw to monkey 2

# Monkey 2:
#   Starting items: 55
#   Operation: new = old + 7
#   Test: divisible by 13
#     If true: throw to monkey 0
#     If false: throw to monkey 7

# Monkey 3:
#   Starting items: 69, 82, 69, 56, 68
#   Operation: new = old + 5
#   Test: divisible by 2
#     If true: throw to monkey 0
#     If false: throw to monkey 2

# Monkey 4:
#   Starting items: 72, 97, 50
#   Operation: new = old + 2
#   Test: divisible by 19
#     If true: throw to monkey 6
#     If false: throw to monkey 5

# Monkey 5:
#   Starting items: 90, 84, 56, 92, 91, 91
#   Operation: new = old * 19
#   Test: divisible by 3
#     If true: throw to monkey 6
#     If false: throw to monkey 1

# Monkey 6:
#   Starting items: 63, 93, 55, 53
#   Operation: new = old * old
#   Test: divisible by 5
#     If true: throw to monkey 3
#     If false: throw to monkey 1

# Monkey 7:
#   Starting items: 50, 61, 52, 58, 86, 68, 97
#   Operation: new = old + 4
#   Test: divisible by 11
#     If true: throw to monkey 5
#     If false: throw to monkey 4
```

```python
import math
```

```python
class monkey_business_example():
    def __init__(self, rounds, worry_loss):
        self.business_list = [0, 0, 0, 0]
        self.worry_loss = worry_loss
        self.rounds = rounds

        self.monkey0_list = [79, 98]
        self.monkey1_list = [54, 65, 75, 74]
        self.monkey2_list = [79, 60, 97]
        self.monkey3_list = [74]

        self.lcm = math.lcm(23, 19, 13, 17)


    def monkey0_func(self, worrylist):
        # print(worrylist)
        for worry in worrylist:
            worry *= 19
            worry = worry % self.lcm
            if worry % 23 == 0:
                self.monkey2_list.append(worry)
            else:
                self.monkey3_list.append(worry)
            self.business_list[0] += 1
        self.monkey0_list = []


    def monkey1_func(self, worrylist):
        for worry in worrylist:
            worry += 6
            worry = worry % self.lcm
            if worry % 19 == 0:
                self.monkey2_list.append(worry)
            else:
                self.monkey0_list.append(worry)
            self.business_list[1] += 1
        self.monkey1_list = []


    def monkey2_func(self, worrylist):
        for worry in worrylist:
            worry = worry * worry
            worry = worry % self.lcm
            if worry % 13 == 0:
                self.monkey1_list.append(worry)
            else:
                self.monkey3_list.append(worry)
            self.business_list[2] += 1
        self.monkey2_list = []


    def monkey3_func(self, worrylist):
        for worry in worrylist:
            worry += 3
            worry = worry % self.lcm
            if worry % 17 == 0:
                self.monkey0_list.append(worry)
            else:
                self.monkey1_list.append(worry)
            self.business_list[3] += 1
        self.monkey3_list = []


    def throwing(self):
        for i in range(self.rounds):
            # print(f'Round {i} Starting.')
            self.monkey0_func(self.monkey0_list)
            self.monkey1_func(self.monkey1_list)
            self.monkey2_func(self.monkey2_list)
            self.monkey3_func(self.monkey3_list)

            # for x in range(8):
            #     self.business_list[x] /= 10
            # print(f'Round {i} Ending.')

        print(self.business_list)
        self.business_list.sort()
        print(f'Total monkey buisness is: {(self.business_list[-1] * self.business_list[-2])}.')
```

```python
test = monkey_business_example(10000, 1)
test.throwing()
```
